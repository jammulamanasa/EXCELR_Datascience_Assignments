```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_excel('/content/drive/MyDrive/Online retail.xlsx',header=None)
df.head()
```

```python
pd.set_option('display.max_rows',None)
df[df.duplicated(keep=False)].head(100)
```

```python
# removing the duplicates
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
# keeping the items purchased in each transaction in a list
df=df.apply(lambda x:x.str.split(','))
df.head()
```

```python
# transformint the data which suits to apply TransactionEncoder
store=[i for i in df[0]]
```

```python
te=TransactionEncoder()
arry=te.fit_transform(store)
```

```python
arry[1]
```

```python
df=pd.DataFrame(arry,columns=te.columns_)
```

```python
df.head()
```

```python
# selecting the frequent items in the transactions with minimum support = 0.009
frequent_itemsets = apriori(df, min_support=0.009, use_colnames=True)
frequent_itemsets
```

```python
# filtering the association rules having lift >1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# mineral water & spaghetti are consequent for every antecedent items
rules[rules.confidence>=0.2]
```

1.	What is lift and why is it important in Association rules?
Lift is a measure that helps evaluate the strength of an association rule by comparing how much more often the items in the rule occur together than if they were statictically independent
if lift>1:positive correlation
if lift<1:negative correlation
if lift=1:X and Y are independent
Impotance:
Lift helps in determining whether the association rule is simply a coincidence or if there is a true relationship between them. Higher lift values suggest assoiations


2.	What is support and Confidence.
support:Support indicates how frequently the itemsets appear in the dataset. It is defined as the proportion of transactions in which the itemset occurs
Support Formula:
[ \text{Support}(X) = \frac{\text{Transactions containing } X}{\text{Total Transactions}} ]
Confidence:
Confidence measures the likelihood of occurrence of the consequent(Y) given that the antecedent(X) has occurred. It helps assess the reliability of the rule
[ \text{Confidence}(X \Rightarrow Y) = \frac{\text{Support}(X \cap Y)}{\text{Support}(X)} ]

3.	What are some limitations or challenges of Association rules mining?
Combinatorial Explosion:
The number of possible itemsets grows exponentially with the number of items, leading to computational inefficiency
Rare Item Problem:
Rare items with small support values may not appear in frequent itemsets even if they have strong associations with other items
Threshold Sensitivity:
Setting the thresholds for support and confidence can be tricky. Too high a threshold may result in missing interesting rules, while too low a threshold might generate too many uninteresting or noisy rules
